```python
#"Use this code for easy RGB to CMY conversion."...
import cv2
image=cv2.imread('/content/Screenshot 2024-04-28 102044.png')
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print("RGB image")
R,G,B=cv2.split(image)
print("R component")
cv2_imshow(R)
print("G component")
cv2_imshow(G)
print("B component")
cv2_imshow(B)
C=1-R
M=1-G
Y=1-B
print("CMY image")
print(" C component")
cv2_imshow(C)
print(" M component")
cv2_imshow(M)
print(" Y component")
cv2_imshow(Y)
```
